**To Create a K-means clustering algorithm to group customers of a retail store based on their purchase history.**

Importing The required libraries


```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

Importing the dataset and naming it as a data frame

```python
df = pd.read_csv('Mall_Customers.csv')
display(df)
```

Split the data into features (X) and target variable (y)

```python
features = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']
X = df[features]
```

Applying K means Algorithm on the given dataset and evaluating the model


```python
kmeans = KMeans(n_clusters=5, random_state=0)
kmeans.fit(X)
df['Cluster'] = kmeans.labels_
```

Plotting the resultant Clusters obtained


```python
plt.figure
plt.scatter(df['Annual Income (k$)'], df['Spending Score (1-100)'], c=df['Cluster'], cmap='viridis', s=50)
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Customer Segmentation based on Income and Spending Score')
plt.show()
```
